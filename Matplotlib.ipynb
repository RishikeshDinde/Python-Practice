# Subplot or multiple plots

```python
a=np.arange(1,6,1)
s=np.arange(1,11,2)
d=np.arange(1,15,3)
f=np.arange(1,21,4)
g=np.arange(1,26,5)
h=np.arange(1,31,6)
j=np.arange(1,36,7)
k=np.arange(1,41,8)
```

```python
a
```

```python
plt.subplot(2,2,1)
plt.plot(a,s)

plt.subplot(2,2,2)
plt.plot(d,f)

plt.subplot(2,2,3)
plt.plot(g,h)

plt.subplot(2,2,4)
plt.plot(j,k)
```

```python
# Title
# supertitle

plt.subplot(2,2,1)
plt.plot(a,s)
plt.title('p1')

plt.subplot(2,2,2)
plt.plot(d,f)
plt.title('p2')

plt.subplot(2,2,3)
plt.plot(g,h)
plt.title('p3')

plt.subplot(2,2,4)
plt.plot(j,k)
plt.title('p4')

plt.suptitle('Multiplots')
```

# Colour map
# It is density of color change from 0 to 100

```python
w=[]

for i in range(0,20):
    x=random.randrange(1,100,2)
    w.append(x)

z=[]
for i in range(0,20):
    x=random.randrange(1,100,2)
    z.append(x)
    
color=[x for x in range(1,100,5)]
plt.scatter(w,z,c=color, cmap='viridis')
```

```python
plt.scatter(w,z,c=color, cmap='viridis')
plt.colorbar()
```

```python
plt.scatter(w,z,c=color, cmap='OrRd_r')
plt.colorbar()
```

# Histogram

# x=np.random.normal(loc, scale, size)
loc : float or array_like of floats
    Mean ("centre") of the distribution.
    
scale : float or array_like of floats
    Standard deviation (spread or "width") of the distribution. Must be
    non-negative.
    
size : int or tuple of ints, optional
    Output shape.  If the given shape is, e.g., ``(m, n, k)``, then
    ``m * n * k`` samples are drawn.  If size is ``None`` (default),
    a single value is returned if ``loc`` and ``scale`` are both scalars.
    Otherwise, ``np.broadcast(loc, scale).size`` samples are drawn.

```python
x=np.random.normal(170,10,250)
x
```

```python
plt.hist(x)
```

```python
plt.hist(x)
```

# Pie chart

```python
x=np.array([35,25,25,15])
plt.pie(x)
```

```python
# Labeling with distance
x=np.array([35,25,25,15])
labels=['Apple','iphone','Jio','idea']
plt.pie(x, labels=labels, labeldistance=.5)
```

```python
# Rotation angle
x=np.array([35,25,25,15])
labels=['Apple','iphone','Jio','idea']
plt.pie(x, labels=labels, labeldistance=.5, startangle=90)
```

```python
# explode
x=np.array([35,25,25,15])
labels=['Apple','iphone','Jio','idea']
ex=[.2,0,0,0]
plt.pie(x, labels=labels, labeldistance=.5, explode=ex)
```

```python
# Shadow
x=np.array([35,25,25,15])
labels=['Apple','iphone','Jio','idea']
plt.pie(x, labels=labels, labeldistance=.5, explode=ex, shadow=True)
```

```python
x=np.array([35,25,25,15])
labels=['Apple','iphone','Jio','idea']

plt.pie(x, labels=labels, labeldistance=.5, explode=ex, shadow=True)
plt.legend()
```
